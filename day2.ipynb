```python
#factors of any number
x = int(input())
n =1
while n<= x:
    if x%n == 0:
        print(n)
    n=n+1
```

```python
#factors counting of any number
x = int(input())
n =1
cnt = 0
while n<= x:
    if x%n == 0:
        cnt = cnt+1
    n=n+1
print(cnt)
```

```python
#Prime number question
x = int(input())
n = 1
cnt = 0
while n<=x:
    if x%n == 0:
        cnt = cnt+1
    n=n+1
if cnt == 2:
    print("Prime Number")
print( "Not a Prime" )
```
